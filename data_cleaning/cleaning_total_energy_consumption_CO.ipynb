```
# Import pandas
import pandas as pd
```

```
# Load the total energy consumption from CO and the MSN codes description.
df = pd.read_excel('/content/drive/My Drive/datasets/data_practicum/total_energy_consumption_CO.xlsx', sheet_name='cleaning')
codes = pd.read_csv('/content/drive/My Drive/datasets/data_practicum/Codes_and_Descriptions_consumption_data.csv', encoding='ISO-8859-1')
```

```
df.head()
```

```
# Turn the MSN, the descriptions, and unit columns into  a list.
code = codes['MSN'].tolist()
desc = codes['Description'].tolist()
unit = codes['Unit'].tolist()
```

```
"""
Set up the format for which the function will iterate through and append the
descriptions and the values of the MSN codes together.
"""
res = {key: {'Description': Description, 'Unit': Unit} for key, Description, Unit in zip(code, desc, unit)}
```

```
res
```

```
# Define both functions for appending descriptions and values.
def desc(x):
  for k,v in res.items():
      if k in x:
        return v['Description']
        
def unit(x):
    for k,v in res.items():
      if k in x:
        return v['Unit']
```

```
# Apply both functions for the description and unit columns to the MSN column.
df['Description'] = df['MSN'].apply(desc)
df['Unit'] = df['MSN'].apply(unit)
```

```
df.head()
```

```
# Export the data into a csv or excel file.
df.to_excel('/content/drive/My Drive/datasets/data_practicum/total_energy_consumption_CO_clean.xlsx', sheet_name='cleaned')
```
