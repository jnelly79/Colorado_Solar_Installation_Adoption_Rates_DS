```
# Import pandas.
import pandas as pd
```

```
# Load solar address data and avg savings income data.
df = pd.read_excel('/content/drive/My Drive/datasets/data_practicum/integrated_random_address_solar_data_final.xlsx')
df1 = pd.read_excel('/content/drive/My Drive/datasets/data_practicum/avg_savings_income.xlsx')
```

```
"""
Create a function avg_savings to locate and find ranges of income for eash
address from the solar address data and pair up their assumed average savings.
"""
def avg_savings(x):
  if int(x) in range(0,2500):
    return float(7164.99)
  if int(x) in range(25001,44999):
    return float(14414.37)
  if int(x) in range(45000,69999):
    return float(19504.92	)
  if int(x) in range(70000,114999):
    return float(13646.37	)
  if int(x) in range(115000,159999):
    return float(67761.00)
```

```
"""
Apply the avg_savings function to the assumed income feature, and create a new
feature called avg_savings_by_income.
"""
df['avg_savings_by_income'] = df['assumed_income'].apply(avg_savings)
```

```
df.head()
```

```
"""
Create a function called affordable_2019 to pair up addresses who can afford
upfront costs in 2019 with their assumed savings.
"""
def affordable_2019(x):
  for i, row in df['upfront_cost_after_incentives'].iteritems():
    if x > float(row):
      return "Y"
    else:
      return "N"
```

```
"""
Apply the function to the the new avg_savings_by_income feature and create a new
feature of Ys and Ns.
"""
df['2019_affordable(YoN)'] = df['avg_savings_by_income'].apply(affordable_2019)
```

```
# Get the value counts of those Ys and Ns.
df['2019_affordable(YoN)'].value_counts()
```

```
# Display the dataframe head.
df.head()
```

```
"""
Like the affordable_2019 function create one for 2020, one that will calculate
the assumed affordability after the incentives implementation for 2020.
"""
def affordable_2020(x):
  for i, row in df['est_new_upfront_cost'].iteritems():
    if x > float(row):
      return "Y"
    else:
      return "N"
```

```
"""
Apply the function affordable_2020 on the avg_savings_by_income feature to 
create a new feature called 2020_incentive_affordable(YoN).
"""
df['2020_incentive_affordable(YoN)'] = df['avg_savings_by_income'].apply(affordable_2020)
```

```
# Display the valued counts of Ys and Ns.
df['2020_incentive_affordable(YoN)'].value_counts()
```
